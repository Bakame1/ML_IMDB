```python
#TASK 1


#RAJOUTER PROCESS RETIRER HTML, MAIL



import pandas as pd
df = pd.read_csv("IMDB Dataset.csv")

def load_imdb_dataset(df):
  """Loads the IMDb dataset from a CSV file.

  Args:
    file_path: Path to the CSV file.

  Returns:
    A tuple of two lists: (train_data, test_data)
  """

  # Assuming the CSV has columns 'review' and 'sentiment'
  train_data = [(review, sentiment) for review, sentiment in zip(df['review'], df['sentiment'])]

  # You can split the data into train and test sets using various methods like random splitting or using a predefined test set.
  # Here's a simple random split:
  import random
  random.shuffle(train_data)
  train_size = int(0.8 * len(train_data))
  train_data, test_data = train_data[:train_size], train_data[train_size:]

  return train_data, test_data

# Example usage:
train_data, test_data = load_imdb_dataset(df)
```

```python
#TASK 2
import nltk
from nltk.tokenize import word_tokenize

def calculate_word_scores(df):
    for index, row in df.iterrows():
        tokens = word_tokenize(row['review'])
        label = row['sentiment']
        
        # Create a dictionary to store word-label pairs
        word_scores = {word: label for word in tokens}
        
        # Update the DataFrame with the dictionary as a new column
        df = df.assign(**word_scores)  # Unpack the dictionary as new column(s)

    return df

# Apply the function to the DataFrame
df = calculate_word_scores(df)
```
